```python
wh_3  <- read.csv("weight-height.csv.gz", header= TRUE)

# i needed to read the csv file and put it in a variable 
# called wh_3

df  <- data.frame(wh_3)

# next i made a data frame with using that variable   

male_H  <- table(round(df[1:5000,2],0))

# then I took only the male from the data frame and 
# rounded their heights and used the table function to get the
# frequency, then put that in a variable called male_H

female_H  <- table(round(df[5001:10000,2],0))

# I got only the female height frequency above

both_H  <- table(round(df$Height,0))

# then I got both the male and female frequency above

plot(male_H, xlab="Male Heights",ylab="frequency") 
plot(female_H, xlab="Female Heights", ylab="frequency", col=c("red"))
plot(both_H, xlab="All Heights", ylab="frequency", col=c("blue"))

#for the next three lines of code, i plotted the male height 
# frequency in black, the female in red, then all in blue

# next I plotted the frecuency of the male heights and labled the x 
# and y axis male heights and frequency, respectively 
```
