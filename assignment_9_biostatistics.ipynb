```r
horse_kicks <- read.csv('HorseKicks.txt', sep="")
horse_kicks

freq <- c()
# i want to create an empty list that stores the frequency data gathered using
# the for loop. how many kicks per year happened across the all cavalry corps

for (x in c(1:length(horse_kicks[,1]))){
    freq <- append(freq,sum(horse_kicks[x,-1]))
    # this remover the first column and gives back the sum of the remainding 
    # columns, for every row
}

freq_kicks <- data.frame(Year=horse_kicks[,1], data.frame(freq))
freq_kicks
# then i made a data frame and with only the year and frequency of kicks that
# occured
#sum(horse_kicks[x,-1])
```

```r
plot(freq_kicks)


shapiro.test(freq_kicks[,2])
# here when using the shapiro test we get a high W which means the data is
# gaussian

qqnorm(freq_kicks[,2],)
abline(a=9.5, b=5, col= "red")
# after printing the qqnorm we see that we have a pretty straight line 
# for best fit
```

```r
# part 1 

# here i am using a function to get the best fit in my data in order
# to make it more gaussian 


x <- horse_kicks[,2]
bxcx <- function(x,lmbd){
    if(abs(lmbd) < 0.05){return(log(x));} else{return((x^lmbd-1)/lmbd)}
    
}

lam  <- seq(-2,2,0.1)
t=list();v=list();cnt=0;
for(i in lam){
    cnt=cnt+1; tt <- shapiro.test(bxcx(x,i));
    t[cnt]=tt$statistic;v[cnt]=i;
}
# the line of best fit looks to be around 0.80

plot(v,t);abline(v=0.80)

r <- bxcx(x, 0.80)

# what this shows me is that i already had data the seemed gaussian
# because from the shapiro test had a high w before hand.
# so the data is a good fit for poisson distribution
qqnorm(r)
hist(r)
#bxcx(horse_kicks[,2])
```

```r
# part 2

mrna <- read.csv("mrnaseq_count_data.csv")
#length(mrna[,1])

k <- c(2:length(mrna))


for (x in k ){
    row <- c(1:length(mrna[,1]))
    
    # this second for loop looks for any 0's in each column and changes it 
    # to the value i want which is the mean of the entire column
    
    for (r in row){ if(mrna[r,x]==0) {mrna[r,x][mrna[r,x]==0]  <- mean(mrna[,x])}}
    
}
mrna
x <- mrna[c(1:5000),2]
bxcx <- function(x,lmbd){
    if(abs(lmbd) < 0.05){return(log(x));} else{return((x^lmbd-1)/lmbd)}
    
}

lam  <- seq(-2,2,0.1)
t=list();v=list();cnt=0;
for(i in lam){
    cnt=cnt+1; tt <- shapiro.test(bxcx(x,i));
    t[cnt]=tt$statistic;v[cnt]=i;
}
# the line of best fit looks to be around 0.80

plot(v,t);abline(v=0.15)
```
